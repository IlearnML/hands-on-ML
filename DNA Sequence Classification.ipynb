그리고 Numerical Method는 진행을 했는데 Embedding layer에서

그리고 Numerical Method 내의 숫자로 encoding은 했는데 string 문자열로 되어 있어서..ㅎㅎ 문제가 좀 있어ㅎㅎ

# Enocoding 1. Numerical Method 

```python
import numpy as np

def nucleotideToDigit(seq):
    digit_seq = []
    seq = list(seq)
    for n in seq:
        if n == 'A' :
            d = 1
        elif n =='C' : 
            d = 2
        elif n == 'G' :
            d = 3
        elif n == 'T' :
            d = 4
        else:
            d = 0
        digit_seq.append(d)
    
    return np.array(digit_seq)
```

```python
import csv
f= open('encoding numerical.csv', 'w', encoding = 'utf-8', newline='')
wr=csv.writer(f)
wr.writerow(["Class", "Sequence"])

with open('Influenza.txt', 'r') as df:

    for seq in df:
        seq=nucleotideToDigit(seq)
        wr.writerow([1,seq])

with open('Covid.txt', 'r') as df:

    for seq in df:
        seq=nucleotideToDigit(seq)
        wr.writerow([2,seq])        

with open('Hepatitis.txt', 'r') as df:

    for seq in df:
        seq=nucleotideToDigit(seq)
        wr.writerow([3,seq])        
        
f.close()
df.close()
```

# Enocoding 2. K-mer Method

```python
KMER_SIZE = 6

def build_kmers(sequence, ksize=KMER_SIZE):
    kmers = []
    n_kmers = len(sequence) - ksize + 1

    for i in range(n_kmers):
        kmer = sequence[i:i + ksize]
        kmers.append(kmer)

    return kmers
```

```python
import csv
f= open('encoding kmer.csv', 'w', encoding = 'utf-8', newline='')
wr=csv.writer(f)
wr.writerow(["Class", "Sequence"])

with open('Influenza.txt', 'r') as df:

    for seq in df:
        kmer = build_kmers(seq)
        wr.writerow([1,kmer])

with open('Covid.txt', 'r') as df:

    for seq in df:
        kmer = build_kmers(seq)
        wr.writerow([2,kmer])        

with open('Hepatitis.txt', 'r') as df:

    for seq in df:
        kmer = build_kmers(seq)
        wr.writerow([3,kmer])        
        
f.close()
df.close()
```

# Embedding Layer

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras import layers
from gensim.models import Word2Vec
import pickle
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('encoding kmer.csv')
```

```python
y=df.Class
X=df.Sequence
print(X.shape, y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=80)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
w2v_model = Word2Vec(X_train, min_count=1, window=5, vector_size=100, workers=2)
```

```python
sample_seq = X_train[0]
sample_seq_kmer= sample_seq[0]
print(f'sample kmer: {sample_seq_kmer}')
print(f'Sample kmer w2v: {w2v_model.wv[sample_seq_kmer]}')
```

승현아 여기에서도 문제가 있어... 예를들면 sample kmer : ATCGAA로 나와야하는데... 그냥 '[' 만 나오네..ㅠㅠ

그래서 잘 돌아가고 있는건지 좀 불투명해 ㅠㅠ 넘 어렵다아 ㅠㅠ

```python
def create_seqs_embedding(w2v_model, seqs):
    seqs_embedding = np.array([np.zeros(100)])
    for seq in seqs: 
        vector = np.zeros(100)
        for word in seq: 
            vector += w2v_model.wv[word]
        number_of_word_in_seq = len(seq)
        normalized_vector = vector / number_of_word_in_seq
        seqs_embedding = np.append(seqs_embedding, [normalized_vector], axis=0)
        
    return seqs_embedding[1: ]
```

```python
seq = 'ACTGGGGGGG'
create_seqs_embedding(w2v_model, seq)
```

```python
X_train_embedding = create_seqs_embedding(w2v_model, X_train)
```

```python
X_test_embedding = create_seqs_embedding(w2v_model, X_test)
```

```python
print(X_train_embedding.shape)
```

# Keras Sequential Layer

```python
model = keras.Sequential([
                             
                             keras.layers.Dense(units=100, activation='relu',input_shape=(100,)),
                             keras.layers.Dense(units=50, activation='relu'),
                             keras.layers.Dropout(rate=0.2),
                             keras.layers.Dense(units=20, activation='relu'),
                             keras.layers.Dense(units=4, activation='softmax')
])
```

```python
model.compile(optimizer='adam', 
              loss=tf.losses.categorical_crossentropy,
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
train_labels = keras.utils.to_categorical(y_train)
train_labels.shape
```

```python
history = model.fit(
    X_train_embedding,
    train_labels,
    epochs=10)
```

```python
test_labels = keras.utils.to_categorical(y_test)
test_labels.shape
```

```python
history2 = model.fit(
    X_test_embedding,
    test_labels,
    epochs=10)
```

```python
import matplotlib.pyplot as plt


plt.plot(history.history['accuracy'])
plt.plot(history2.history['accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['T rain','T est'], loc='lower right')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history2.history['loss'])
plt.title('Model loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(['T rain','T est'], loc='upper right')
plt.show()
```
